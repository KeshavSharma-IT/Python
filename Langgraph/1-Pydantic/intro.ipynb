Pydantic models are the foundation of data validation in python. They use Pythin type annotation to define the structure and validate data at runtime. Here's detailed exploration of basic model crearion with several examples.


```python
from pydantic import BaseModel
```

```python
class Person(BaseModel):
    name:str
    age:int
    city:str

person=Person(name="Keshav",age=35,city="Ghaziabad");


print(person);
```

```python
person1=Person(name="Keshav",age=35,city=43);
print(person1);
```

#2 MOdel with Optional Fields

```python
from typing import  Optional
class Employee(BaseModel):
    id:int
    name:str
    department:str
    salary:Optional[float]=None
    is_active:Optional[bool]=True
```

```python
emp1=Employee(id=1,name="Keshav",department="IT")
print(emp1);
```

```python
emp2=Employee(id=1,name="Keshav",department="IT",salary=6000,is_active=True)
print(emp2);
```
